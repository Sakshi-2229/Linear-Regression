```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv(r'E:/Btech/Linear_single_variable.csv')
```

```python
df
```

```python
sns.lmplot(x='Age',y='Premium',data=df)
```

```python
reg=linear_model.LinearRegression()
```

```python
#fit is used for training the data
reg.fit(df[['Age']],df['Premium'])  # we have used two brackets because the data will be in 2D format
```

```python
#as now our model is trained it is ready for predictions.
# for predictions follow below steps
reg.predict([[21]])   #21 is the age
```

we can see in dataset for age 25, premium is 18000. so for age 21 it should be less than 18000. so here the predicted premium for age 21 is 13880

```python
#for age 50
reg.predict([[50]])
```

now here we can see that the age in dataset is till 45. but here we have predicted age=50. so as in dataset for age=45 the premium is 55000. so for age=50 premium should be above 55000. so here we have predicted the age=50, premium is 65500. so we can say that our model is giving correct predictions.

we will validate y=mx+c
here our linear model will give you slope. slope means coefficient

```python
reg.coef_
```

here slope is 1780
